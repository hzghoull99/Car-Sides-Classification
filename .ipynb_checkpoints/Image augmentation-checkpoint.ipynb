```python
import cv2
import imgaug.augmenters as iaa
import glob
import os
import imgaug
```

```python
images=[]
image_path=glob.glob(r"C:\Users\Hamza\Desktop\training_data1\Left view\*.jpg")
```

```python
image_path
```

```python
len(image_path)
```

```python
for path in image_path:
    img=cv2.imread(path)
    img=cv2.resize(img,(512,512))
    images.append(img)
```

```python
for i in range(8):
   
    augmentation=iaa.Sequential([
    
        #iaa.Flipud(0.8), # vertical flip
        #iaa.Fliplr(.5), # horzontal flip (we can apply it on "front" and "rear")
        iaa.Affine(scale=[0.9,1.1],translate_percent=(-0.12,0.12),
                 rotate =[-5,5],mode='edge',shear={"x": (-5, 7), "y": (-5, 8)}),
        iaa.MultiplyAndAddToBrightness(mul=(0.7, 1.7), add=(-10, 12)),
        iaa.MultiplySaturation((0.7, 1.4)),
        iaa.MultiplyHue((0.9, 1.1)),
        iaa.LinearContrast((0.5, 1.6)),
        iaa.Sometimes(0.2,iaa.GaussianBlur(sigma=(0.0, 2.5))),
        iaa.Sometimes(0.09,iaa.Snowflakes(flake_size=(0.1, 0.2), speed=(0.01, 0.03))),
        iaa.Sometimes(0.2,iaa.Canny(alpha=(0.0, 0.3))) 
    ])
    
    augmented_images=augmentation(images=images)
    
#    for img in augmented_images:
#        cv2.imshow("Images",img)
#        cv2.waitKey(0)
#        cv2.destroyAllWindows()
        
    for j in range (len(image_path)):
        cv2.imwrite(r'C:\Users\Hamza\Desktop\imgaug\Left view aug\aug{}_{}.jpg'.format(i,j),augmented_images[j])    
```
